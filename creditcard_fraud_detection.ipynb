Undersampling
build a sample dataset containing similar distribution of normal transactions and fraudlent transactions


```python
#number of fraudlent transactions is 45
legit_sample=legit.sample(n=45)
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
print(new_dataset)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['is_fraud'].value_counts()
```

```python
new_dataset.groupby('is_fraud').mean()
```

```python
x=new_dataset.drop(columns='is_fraud',axis=1)
y=new_dataset['is_fraud']

x.head()
```

```python
x["trans_date_trans_time"] = pd.to_datetime(x["trans_date_trans_time"])
x["dob"] = pd.to_datetime(x["dob"])
x
```

```python
x.drop(columns=['Unnamed: 0','cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],inplace=True)
x
```

```python
#Drop all rows that contain missing values
x = x.reset_index(drop=True)
x
```

```python
encoder = LabelEncoder()
x["merchant"] = encoder.fit_transform(x["merchant"])
x["category"] = encoder.fit_transform(x["category"])
x["gender"] = encoder.fit_transform(x["gender"])
x["job"] = encoder.fit_transform(x["job"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
x.shape,x_train.shape,x_test.shape
```

model creation

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#accuracy score
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print(training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print(test_data_accuracy)
```
